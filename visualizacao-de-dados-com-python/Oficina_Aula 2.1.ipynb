**01) Importe as bibliotecas pandas, matplotlib, seaborn e use o comando mágico %matplotlib inline para exibir as visualizações no notebook. Carregue os dados em um DataFrame. Os dados estão disponibilizados no link: dados. Você pode carregar os dados pela URI ou salvar em sua máquina e carregá-los localmente.**

```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns

%matplotlib inline
```

```python
filmes_uri = 'https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro03/oficina/arquivos/dadosoficina03.txt'
filmes = pd.read_json(filmes_uri)
filmes.head(30)
```

**02) Crie uma visualização que mostre o orçamento de todos os filmes do conjunto de dados. Qual o filme com maior orçamento e com menor orçamento?**

```python
plt.figure(figsize=(15,8))
plt.title('Filme x Orçamento (2007 - 2011)')
sns.barplot(x='Budget_M', y='Film', color = 'salmon',order=filmes.sort_values('Budget_M').Film, data=filmes)
plt.show()
```

O filme com maior orçamento é o: Pirates of the Caribbean: At World's End

O filme com menor orçamento é o: The Twilight Saga: New Moon

**03) Crie uma visualização que mostre a bilheteria de todos os filmes do conjunto de dados. Qual o filme com maior bilheteria e com menor bilheteria?**

```python
plt.figure(figsize=(15,8))
plt.title('Filme x Bilheteria  (2007 - 2011)')
sns.barplot(x='Worldwide_Gross_M', y='Film', color = 'violet',order=filmes.sort_values('Worldwide_Gross_M').Film, data=filmes)
plt.show()
```

O filme com maior bilheteria é o: Avatar

O filme com menor bilheteria é o: Ratatouille

**04) Crie uma visualização que mostre a quantidade de filmes por gênero em cada ano. Em que anos foram feitos menos filmes de ação? E em que ano foram feitos menos filmes de animação?**

```python
sns.set(style="whitegrid")
plt.figure(figsize=(15,6))
plt.title('Qtde de filmes por gênero (2007 - 2011)')
sns.countplot(x="Year", hue = 'Genre', palette = "pastel",data=filmes)
plt.show()
```

Foram feitos menos filmes de ação em 2008 e 2010.

Foram feitos menos filmes de animação em 2008 e 2011.

**05) Crie uma visualização que mostre a média de orçamento de filmes por gênero em cada ano. Em que ano os filmes de ação tiveram maior média de orçamento? E em que ano os filmes de animação tiveram a maior média de orçamento?**

```python
sns.set(style="whitegrid")
plt.figure(figsize=(15,8))
plt.title('Média de orçamento por gênero de filme (2007 - 2011)')
sns.barplot(x="Year", y = 'Budget_M', hue= 'Genre', palette = "pastel", ci = 'sd', data=filmes)
plt.show()
```

Os filmes de ação tiveram maior média de orçamento em 2007.

Os filmes de animação tiveram maior média de orçamento em 2008.
