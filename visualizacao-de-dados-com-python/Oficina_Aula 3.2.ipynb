```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
tips = sns.load_dataset('mpg')
tips.head()
```

**02) Crie uma visualização que mostre a relação entre a aceleração (‘acceleration’) e o consumo de combustível (‘mpg’) dos veículos, em que seja possível diferenciar o país de origem (‘origin’). Sabendo disso, existe uma correlação entre a aceleração e o consumo? De que tipo? A correlação é a mesma para os diferentes países de origem? Você pode escolher a técnica que julgar mais adequada!**

**Matriz de correlação: para verificar as correlações em cada país**

```python
correlacao_usa = tips[tips['origin'] == 'usa'].corr()
correlacao_usa
```

```python
correlacao_japan = tips[tips['origin'] == 'japan'].corr()
correlacao_japan
```

```python
correlacao_europe = tips[tips['origin'] == 'europe'].corr()
correlacao_europe
```

```python
sns.relplot(x="mpg", y="acceleration", col = 'origin', color = 'green', data=tips)
```

Há correlação entre aceleração e consumo de combustível.

A correlação entre a aceleração e o consumo de combustível é positiva e mais forte em usa e japan.

**03) Crie uma visualização que mostre a relação entre a aceleração (‘acceleration’) e o consumo de combustível (‘mpg’) dos veículos, em que seja possível visualizar o ano do modelo (model_year). Existe alguma relação entre as três variáveis?**

```python
plt.figure(figsize=(15,8))
plt.title('Acceleração x Consumo de combustível x Ano do modelo')
sns.scatterplot(x="mpg", y="acceleration", size = 'model_year', sizes=(15, 200), data=tips)
plt.show()
```

Os modelos de carro a partir dos anos 76 tem uma correlação positiva em relação à aceleração x o consumo de combustível. Os pontos menores tendem a ter menor aceleração e consumo de combustível, e os pontos maiores, tentem a ter maior aceleração e consumo de combustível.
