```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('houses.txt')
df.head()
```

Dica da monitora dem fazer o histograma de forma diferente:(por isso importei o numpy)

```python
#Plotando a distribuição de aluguéis dos imóveis
plt.figure(figsize=(20, 10))
plt.style.use('seaborn-darkgrid')
plt.title('Distribuição de aluguéis dos imóveis')
#Criando variável para controlar o número de bins, dividindo os dados de 0 ao valor máximo, em um certo número de intervalos
number_bins = np.linspace(0,df['rent amount (R$)'].max(),20)
sns.histplot(df['rent amount (R$)'],bins=number_bins,kde=True)
plt.xticks(number_bins) #rótulos do eixo x
plt.show()
```

**02) Crie uma visualização que mostre a distribuição dos valores de aluguel, ‘rent amount (R$)’, dos imóveis. Pode utilizar o gráfico que você julgar mais adequado. Qual faixa de valores de aluguel é mais comum?**

```python
plt.figure(figsize=(10, 5))
plt.style.use('Solarize_Light2')
plt.title('Distribuição dos valores (R$) de aluguel')
plt.hist(df['rent amount (R$)'],bins=5, color = 'violet', edgecolor='gray')
plt.show()
```

Há um pouco mais de 7000 imóveis em que a faixa do valor do aluguel vai até 5000 (R$). Percebemos essa ocorrência na primeira barra do histograma.

**03) Crie uma visualização que mostre a distribuição dos valores de seguro contra incêndio, 'fire insurance (R$)', dos imóveis. Existem valores incomuns?**

```python
plt.figure(figsize=(10, 5))
plt.style.use('Solarize_Light2')
plt.title('Distribuição dos valores (R$) de seguro contra incêndio dos imóveis')
plt.hist(df['fire insurance (R$)'],bins=7, color = 'orange', edgecolor='red')
plt.show()
```

Há um pouco mais de 6000 imóveis em que a faixa do valor do seguro contra incêndio vai até 50 (R$). Percebemos essa ocorrência na primeira barra do histograma. A partir daí, as quantidades vão caindo e os valores de seguros mais elevados são mais raros.
