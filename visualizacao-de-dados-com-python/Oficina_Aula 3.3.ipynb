```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
vendas = pd.read_csv('https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro07/oficina/arquivos/dadosoficina07.txt')
vendas.head()
```

**02) Para o segundo e último passo, crie uma visualização que mostre a série temporal das vendas ao longo dos meses. Você pode, por exemplo, se perguntar se existe alguma tendência ou alguma periodicidade, OK?**

```python
plt.figure(figsize=(18,8))
plt.title('Variação das Vendas ao longo do tempo')
sns.lineplot(x="Month", y="Sales", color ='orange', linewidth = 1, marker = 'o', ms = 8, data=vendas)
plt.show()
```

Observando o gráfico do valor das vendas, constata-se que esse valor em alguns períodos apresenta queda, mas ele está aumentando consideralvelmente ao longo dos meses.
